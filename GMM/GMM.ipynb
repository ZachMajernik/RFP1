```python
import pandas as pd
from sklearn.mixture import GaussianMixture
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("customer_profiles.csv")

# Select relevant features (for example, "age" and "income")
selected_features = df[['age','income']]

# Train the Gaussian Mixture Model
gmm = GaussianMixture(n_components=4, random_state=0)
labels = gmm.fit_predict(selected_features)

# Add cluster labels to the DataFrame
df['cluster'] = labels

# Visualize the clusters
sns.scatterplot(x='age', y='income', hue='cluster', data=df, palette='viridis')
plt.title('Gaussian Mixture Model Clusters')
plt.show()
```

```python
import pandas as pd
from sklearn.mixture import GaussianMixture
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("customer_profiles.csv")

# Select relevant features (for example, "age" and "income")
selected_features = df[['age','income']]

# Train the Gaussian Mixture Model
gmm = GaussianMixture(n_components=4, random_state=0)
labels = gmm.fit_predict(selected_features)

# Add cluster labels to the DataFrame
df['cluster'] = labels

# Visualize the clusters with a different palette
sns.scatterplot(x='age', y='income', hue='cluster', data=df, palette='Set1')  # Change the palette to 'Set1'
plt.title('Gaussian Mixture Model Clusters')
plt.show()
```
